```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
prices = pd.read_csv('./Nat_Gas.csv', parse_dates=['Dates'], dayfirst=True)
prices.set_index('Dates', inplace=True)
```

```python
display(prices)
```

```python
def contract(injection, withdraw, prices, rate, volume, transportation, storage_cost):
    #check whether the contract is feasible i.e. not violate the volume
    prices['Injection'] = np.where(prices.index.isin(injection), rate, 0)
    prices['Withdraw'] = np.where(prices.index.isin(withdraw), rate, 0)
    prices['Current Amount'] = prices['Injection'].cumsum() - prices['Withdraw'].cumsum()
    if (prices['Current Amount'] >= 0).all() and (prices['Current Amount'] <= volume).all():
        # total storage cost
        months = (prices['Current Amount'] != 0).sum()
        total_storage_cost = storage_cost * months
        # marginals
        marginal = (prices['Withdraw'] * prices['Prices']).sum() - (prices['Injection'] * prices['Prices']).sum()
        # transportation
        trans_cost = transportation * (len(injection) + len(withdraw))
        # total cost
        total = marginal - total_storage_cost - trans_cost
        return(total)
    else:
        print('infeasible contract!')
```

```python
injection = pd.to_datetime(['2021-05-31', '2022-04-30', '2023-07-31'])
withdraw = pd.to_datetime(['2022-01-31', '2022-12-31', '2023-12-31'])
rate = 5000
volume = 15000
transportation = 100
storage_cost = 500
```

```python
contract(injection, withdraw, prices, rate, volume, transportation, storage_cost)
```
